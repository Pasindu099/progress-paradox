```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("total-oda-for-infrastructure-by-recipient.csv")
oda_col = df.columns[3]
df = df.rename(columns={oda_col: "ODA"})

region_map = {
    "Africa (UN)": "Africa",
    "Asia (UN)": "Asia",
    "Europe (UN)": "Europe",
    "Latin America and the Caribbean (UN SDG)": "Americas",
    "Oceania (UN SDG)": "Oceania",
}

regions_df = df[df["Entity"].isin(region_map.keys())].copy()
regions_df["Region"] = regions_df["Entity"].map(region_map)

# Calculate percent share per year
regions_df["TotalODA"] = regions_df.groupby("Year")["ODA"].transform("sum")
regions_df["Share"] = regions_df["ODA"] / regions_df["TotalODA"]

fig = px.area(
    regions_df,
    x="Year",
    y="Share",
    color="Region",
    title="Share of Global Infrastructure ODA by Region ",
)

fig.update_layout(yaxis_tickformat=".0%")

fig.show()
```
